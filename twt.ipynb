```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def directWave(offset,v_0):
    return 1000*(offset/v_0)
```

```python
def reflectedWave(offset,depth,v_0):
    return 1000*((2/v_0)*(np.sqrt(depth**2+(offset/2)**2)))
```

```python
def refractedWave(offset,depth,v_0,v_1):
    return 1000*(offset/v_1 + 2*depth*np.sqrt(v_1**2-v_0**2)/(v_0*v_1))
```

```python
offset = np.linspace(0,600,100)
depth = 50
v_0 = 1500
v_1  = 2000
```

```python
directTime = directWave(offset,v_0)
reflectedTime = reflectedWave(offset, depth, v_0)
refractedTime = refractedWave(offset,depth,v_0,v_1)
```

```python
plt.plot(offset,directTime,label='Direct Wave')
plt.plot(offset,reflectedTime,label='Reflected Wave')
plt.plot(offset,refractedTime,label='Refracted Wave')
plt.legend()
plt.title('Two way Travel (TWT) time')
plt.xlabel('Offset')
plt.ylabel('TWT (ms)')
plt.show()
```
